# Question 5 : Parmi les 100 films les mieux notés, quel pourcentage sont américains ? Français ?


```python
percent = movies.find().sort('score',-1).limit(100)

fr = 0
usa = 0
for el in percent:
    if 'France' in el["origine"]:
        fr+=1
    elif 'United States' in el['origine']:
        usa+=1

print('le pourcentage de film francais est ' + str(fr) + '%'  ' et le pourcentage de film american est ' + str(usa) +'%')
```

# Question 6 : Quel est la durée moyenne d’un film en fonction du genre ?


```python

all_movies = movies.find()
genre = []
for elt in all_movies:
    if elt['genre'] not in genre :
        genre.append(elt['genre'])

total_filmbygenre = {el:0 for el in genre}
nb_filmbygenre = {el:0 for el in genre}



fdp = movies.find()


for elt in fdp:
    total_filmbygenre[elt['genre']] += (elt['duree'])
    nb_filmbygenre[elt['genre']] += 1

mean_filmbygenre = dict((k, float(total_filmbygenre[k]) / nb_filmbygenre[k]) for k in total_filmbygenre)


```
